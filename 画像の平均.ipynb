##### 平均画像生成（コメントあり）

```python
import numpy as np
import cv2
import os

# 関数定義（画像のサイズを取得する）
def minsize(s):
    
    count = 0
    # ファイル分ループ
    for i in os.listdir(dir):
        # 画像読み込み
        im = cv2.imread(dir + i)
        
        # もし s(9999) より画像の縦サイズが小さかったら
        if s > int(im.shape[0]):
            # カウントにプラス１
            count += 1
            
            # もしカウントが２ならば動く処理
            if count == 2:
                print("画像のサイズが統一されていません")
                print("処理を終了します")
                # 処理を終了させる
                sys.exit()
            # s に縦サイズを代入
            s = int(im.shape[0])
            
    # 結果を返す
    return s

# ファイル数を取得するフォルダの指定
dir = '../Production_exercise/cv/resize_ok/'
# ファイル数を取得 num に格納
num = sum(os.path.isfile(os.path.join(dir, name)) for name in os.listdir(dir))
# 関数呼び出し（値は9999）
size = minsize(9999)
# size * size の総画素数を pnum に格納
pnum = size * size
# 縦　ファイル数 num 　：　横　総画素数 pnum 　の二次元配列を生成
allarray = np.empty((num, pnum))
# 計算の答え置き
pixel = 0
# 平均画素数格納用の長さ pnum の一次元配列を生成
ave = np.empty(pnum)

# ファイル数分ループ
for i in range(num):
    # 画像をグレースケール化 ＆ 読み込み
    img = cv2.imread(dir + "resize_face" + str(i) + ".jpg",cv2.IMREAD_GRAYSCALE)
    # 多次元配列である img を　reshape(-1) で一次元配列に変換
    # 二次元配列の i 番目に格納   
    allarray[i, :]  =  img.reshape(-1)

    
    
    
# 総画素数文ループ
for i in range(pnum):
    
    # ファイル分ループ
    for j in range(num):
        # 二次元配列の縦の行の値のすべてを加算 pixel に格納
        pixel = pixel + allarray[j,i]
    
    # 加算されているかの確認
    print(i,pixel,end = "\t")
    
    # pixel をファイル数で割る（小数点切り捨て）
    pixel //= num
    # 一次元配列の i 番目に格納
    ave[i] = pixel
    
    # 割れているかの確認
    print(i,ave[i])
    
# 一次元配列を size * size の多次元配列 array に格納
array = np.reshape(ave, (size, size))
# 保存
cv2.imwrite('ave_favce.jpg', array)
```
